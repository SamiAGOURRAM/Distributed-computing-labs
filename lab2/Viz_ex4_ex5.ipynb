```python
import matplotlib.pyplot as plt

threads = [1, 2, 4, 8, 16]
times = [1.263831, 0.852549, 0.839511, 0.791790, 0.821714]
speedup = [times[0] / t for t in times]
efficiency = [s / p for s, p in zip(speedup, threads)]

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(threads, speedup, marker='o')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup')
plt.title('Speedup vs Threads')

plt.subplot(1, 2, 2)
plt.plot(threads, efficiency, marker='o')
plt.xlabel('Number of Threads')
plt.ylabel('Efficiency')
plt.title('Efficiency vs Threads')

plt.tight_layout()
plt.show()
```

# Exercice 5

```python
import matplotlib.pyplot as plt

# Thread counts corresponding to the different runs
threads = [1, 2, 4, 8, 16]

# CPU times in seconds from your output:
# For 1, 2, 4, 8, 16 threads respectively.
cpu_times = [1.340e-03, 1.067e-03, 4.086e-03, 1.138e-02, 2.174e-02]

# Compute speedup: speedup(p) = T(1) / T(p)
baseline = cpu_times[0]
speedup = [baseline / t for t in cpu_times]

# Compute efficiency: efficiency(p) = speedup(p) / p
efficiency = [s / p for s, p in zip(speedup, threads)]

# Print the computed values for reference
print("Threads:", threads)
print("CPU Times (s):", cpu_times)
print("Speedup:", speedup)
print("Efficiency:", efficiency)

# Plotting speedup and efficiency
plt.figure(figsize=(12, 5))

# Speedup plot
plt.subplot(1, 2, 1)
plt.plot(threads, speedup, marker='o', linestyle='-', color='blue')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup (T1 / T_p)')
plt.title('Speedup vs. Number of Threads')
plt.xticks(threads)
plt.grid(True)

# Efficiency plot
plt.subplot(1, 2, 2)
plt.plot(threads, efficiency, marker='o', linestyle='-', color='green')
plt.xlabel('Number of Threads')
plt.ylabel('Efficiency (Speedup / Threads)')
plt.title('Efficiency vs. Number of Threads')
plt.xticks(threads)
plt.grid(True)

plt.tight_layout()
plt.show()
```
